```python
import pandas as pd
```

```python
best = pd.read_csv('data/best_score.csv', index_col=0)
best.head()
```

# Best 5점

```python
best_5 = best[best['best_score'] == 5]
best_5.head()
```

```python
best_5 = best_5[['cell_id', 'cell_ycrd', 'cell_xcrd']]
best_5.head()
```

```python
best_5_dict = best_5.set_index(['cell_id']).T.to_dict('list')
best_5_dict
```

```python
office = pd.read_csv('res/관공서현황.csv', index_col=0)
office.head()
```

```python
office = office[['소재지명', '위도', '경도']]
office.head()
```

```python
office_dict = office.set_index(['소재지명']).T.to_dict('list')
office_dict
```

```python
from haversine import haversine
```

```python
df_all = pd.DataFrame(columns=["best_name", "office_name", "거리차이", "score"])

for i in range(0, len(best_5_dict)) :
    best_nm = list(best_5_dict.keys())[i]
    best_xy = list(best_5_dict.items())[i][1]

    for j in range(0, len(office_dict)) :
        office_nm = list(office_dict.keys())[j]
        office_xy = list(office_dict.items())[j][1]

        ## 최종 거리 계산 결과값
        result = haversine(best_xy, office_xy, unit='m')

        score = 0
        if result <= 200:
            score = 5
        elif 200 < result <= 400:
            score = 4
        elif 400 < result <= 600:
            score = 3
        elif 600 < result <= 800:
            score = 2
        elif 800 < result <= 1000:
            score = 1
        df_all = pd.concat([df_all, 
                        pd.DataFrame({'best_name':[best_nm],'office_name':[office_nm], '거리차이':[result], 'score':[score]})], ignore_index=True)
df_all
```

```python
df_all_5 = df_all.groupby('office_name')['score'].sum()
df_all_5 = df_all_5.reset_index()
df_all_5 = df_all_5.rename(columns = {'score' : '5_score'})
df_all_5
```
