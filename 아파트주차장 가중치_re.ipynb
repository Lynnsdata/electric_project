```python
import pandas as pd
```

```python
best = pd.read_csv('data/best_score.csv', index_col=0)
best.head()
```

# Best 5Ï†ê

```python
best_5 = best[best['best_score'] == 5]
best_5.head()
```

```python
best_5 = best_5[['cell_id', 'cell_ycrd', 'cell_xcrd']]
best_5.head()
```

```python
best_5_dict = best_5.set_index(['cell_id']).T.to_dict('list')
best_5_dict
```

```python
apt_parking_best = pd.read_csv('res/ÏïÑÌååÌä∏Ï£ºÏ∞®Ïû•.csv', index_col=0)
apt_parking_best
```

```python
apt_parking_best = apt_parking_best[['ÏÜåÏû¨ÏßÄÎ™Ö', 'ÏúÑÎèÑ', 'Í≤ΩÎèÑ']]
apt_parking_best
```

```python
apt_parking_best_dict = apt_parking_best.set_index(['ÏÜåÏû¨ÏßÄÎ™Ö']).T.to_dict('list')
apt_parking_best_dict
```

```python
from haversine import haversine
```

```python
df_all = pd.DataFrame(columns=["best_name", "apt_parking_name", "Í±∞Î¶¨Ï∞®Ïù¥", "score"])

for i in range(0, len(best_5_dict)) :
    best_nm = list(best_5_dict.keys())[i]
    best_xy = list(best_5_dict.items())[i][1]

    for j in range(0, len(apt_parking_best_dict)) :
        apt_parking_nm = list(apt_parking_best_dict.keys())[j]
        apt_parking_xy = list(apt_parking_best_dict.items())[j][1]

        ## ÏµúÏ¢Ö Í±∞Î¶¨ Í≥ÑÏÇ∞ Í≤∞Í≥ºÍ∞í
        result = haversine(best_xy, apt_parking_xy, unit='m')

        score = 0
        if result <= 200:
            score = 5
        elif 200 < result <= 400:
            score = 4
        elif 400 < result <= 600:
            score = 3
        elif 600 < result <= 800:
            score = 2
        elif 800 < result <= 1000:
            score = 1
        df_all = pd.concat([df_all, 
                        pd.DataFrame({'best_name':[best_nm],'apt_parking_name':[apt_parking_nm], 'Í±∞Î¶¨Ï∞®Ïù¥':[result], 'score':[score]})], ignore_index=True)
df_all
```

```python
df_all_5 = df_all[df_all['score'] == 5]
df_all_5
```

```python
df_all_5 = df_all_5.groupby('apt_parking_name')['score'].sum()
df_all_5 = df_all_5.reset_index()
df_all_5 = df_all_5.rename(columns = {'score' : '5_score'})
df_all_5
```
