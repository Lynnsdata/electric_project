```python
import pandas as pd
import folium
import json
```

```python
path = 'data/HangJeongDong_ver20220401.geojson'
```

```python
rfile = open(path, 'r', encoding='utf8').read()
jsonfile =json.loads(rfile)

json_yongin = {'type': 'FeatureCollection'}
json_pick = []
json_dong = []
for item in jsonfile['features']:
    # print(item)
    if item['properties']['sggnm']=='용인시처인구' or item['properties']['sggnm']=='용인시수지구' or item['properties']['sggnm']=='용인시기흥구' :
        # print(item)
        dong = item['properties']['adm_nm'].split()[-1]

        # print(dong)
        item['id'] = dong
        # print(item)
        json_dong.append(dong)
        json_pick.append(item)
json_yongin['features']=json_pick
```

# best

```python
best = pd.read_csv('data/best_score.csv', index_col=0)
best.head()
```

```python
best_5 = best[best['best_score'] == 5]
best_5.head()
```

```python
map = folium.Map((37.240608, 127.1772935), zoom_start=10)

folium.GeoJson(json_yongin).add_to(map)

for i in best_5.index:
  sub_lat = best_5.loc[i, 'cell_ycrd']
  sub_long = best_5.loc[i, 'cell_xcrd']
  
  title = f"{best_5.loc[i, 'cell_ycrd']} - {best_5.loc[i, 'cell_xcrd']}"
  
  folium.Marker(location=[sub_lat, sub_long],
                      icon=folium.Icon(color='pink'),
                      tooltip=title).add_to(map)

map
```
