```python
import pandas as pd
```

```python
import os
base_dir = './res/'
dir = os.path.join(base_dir, '')
dir
```

```python
data = os.listdir(dir)
data
```

```python
df_all = pd.DataFrame(columns=['소재지명', '도로명', '위도', '경도', '시설구분'])
df_all
```

```python
for i in range(len(data)):
    df_all = pd.concat([df_all, (pd.read_csv(dir+data[i]))], ignore_index=True)
df_all
```

```python
df_all = df_all.drop(['Unnamed: 0'], axis=1)
df_all
```

```python
df_all.info()
```

```python
df_all = df_all.drop_duplicates(['도로명'])
df_all
```

```python
df_all = df_all.reset_index(drop=True)
df_all
```

```python
for i in range(len(df_all)):
    df_all['도로명'][i] = df_all['도로명'][i].rstrip()
df_all
```

```python
df_all['score'] = 0

for i in range(len(df_all)):
    if (df_all.loc[i, '시설구분'] == '숙박업소') | (df_all.loc[i, '시설구분']  == '아파트주차장') | (df_all.loc[i,'시설구분'] == '관공서'):
        df_all['score'][i]  = 4
    elif (df_all.loc[i,'시설구분'] == '골프장') | (df_all.loc[i,'시설구분'] == '시장및마트') | (df_all.loc[i,'시설구분']== '대규모점포') | (df_all.loc[i,'시설구분']  == '응급의료기관'):
        df_all['score'][i]  = 3
    elif (df_all.loc[i,'시설구분'] == '영화상영관') | (df_all.loc[i,'시설구분'] == '공연장') | (df_all.loc[i,'시설구분'] == '박물관및미술관'):
        df_all['score'][i] = 2
    elif (df_all.loc[i,'시설구분'] == '음식점') | (df_all.loc[i,'시설구분'] == '의원') | (df_all.loc[i,'시설구분'] == '일반음식점'):
        df_all['score'][i] = 1
    else:
        df_all['score'][i] = 0
df_all
```
