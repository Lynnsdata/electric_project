```python
import pandas as pd
```

# 컬럼 rename

```python
# 관공서
office = pd.read_csv(r'C:\Users\admin\Desktop\electric_charging_hj\res\관공서현황.csv', index_col=0)
office.head()
```

```python
office.rename(columns={'소재지명':'place'}, inplace=True)
office.head()
```

```python
office_df = pd.read_csv(r'C:\Users\admin\Desktop\electric_charging_hj\res\관공서점수.csv', index_col=0)
office_df.head()
```

```python
office_df.rename(columns={'office_name':'place'}, inplace=True)
office_df.head()
```

```python
office_df_merge = pd.merge(office, office_df, on='place', how='inner')
office_df_merge.head()
```

```python
# 숙박업체
hotel = pd.read_csv(r'C:\Users\admin\Desktop\electric_charging_hj\res\숙박업소현황.csv', index_col=0)
hotel.head()
```

```python
hotel.rename(columns={'소재지명':'place'}, inplace=True)
hotel.head()
```

```python
hotel_df = pd.read_csv(r'C:\Users\admin\Desktop\electric_charging_hj\res\숙박업체점수.csv', index_col=0)
hotel_df
```

```python
hotel_df.rename(columns={'hotel_name':'place'}, inplace=True)
hotel_df.head()
```

```python
hotel_df_merge = pd.merge(hotel, hotel_df, on='place', how='inner')
hotel_df_merge.head()
```

```python
# 아파트 주차장
aptP = pd.read_csv(r'C:\Users\admin\Desktop\electric_charging_hj\res\아파트주차장.csv', index_col=0)
aptP.head()
```

```python
aptP.rename(columns={'소재지명':'place'}, inplace=True)
aptP
```

```python
aptP_df = pd.read_csv(r'C:\Users\admin\Desktop\electric_charging_hj\res\아파트주차장점수.csv', index_col=0)
aptP_df
```

```python
aptP_df.rename(columns={'apt_parking_name':'place'}, inplace=True)
aptP_df
```

```python
aptP_df_merge = pd.merge(aptP, aptP_df, on='place', how='inner')
aptP_df_merge
```

# 최종 concat

```python
df_score_all = pd.concat([office_df_merge, hotel_df_merge, aptP_df_merge], ignore_index=True)
df_score_all
```

```python
df_score_all = df_score_all.drop_duplicates(['place'])
df_score_all
```

```python
df_score_all = df_score_all.sort_values(by='Total' ,ascending=False)
df_score_all
```

```python
df_score_all.reset_index(drop=True, inplace=True)
df_score_all
```

```python
df_score_all.to_csv('./res/최종가중치_merge.csv')
```
