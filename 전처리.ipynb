# 모범음식점

```python
restaurant = pd.read_csv('data/경기도 용인시_모범음식점 지정 정보_20220721.csv', encoding='cp949')
restaurant.head()
```

```python
restaurant.info()
```

```python
restaurant['소재지(도로명)'].str.split('(')[1]
```

```python
restaurant['소재지(도로명)'].str.split('(')[5][0]
```

```python
restaurant['소재지(도로명)'].str.split('(')[5][0].split(',')[0]
```

```python
restaurant['도로명'] = 0
for i in range(len(restaurant)):
    restaurant['도로명'][i] = restaurant['소재지(도로명)'].str.split('(')[i][0].split(',')[0].rstrip()
restaurant.head(3)
```

```python
restaurant = restaurant.rename(columns={'업소명' : '소재지명'})
restaurant.head(3)
```

```python
restaurant['시설구분'] = '음식점'
restaurant.head(3)
```

```python
addresses = []

for i in restaurant.index:
  addresses.append(restaurant['도로명'][i])

addresses
```

```python
geo_local = Nominatim(user_agent='South Korea')

def geocoding(addresses):
    try:
        geo = geo_local.geocode(addresses)
        x_y = [geo.latitude, geo.longitude]
        return x_y

    except:
        return [0,0]
```

```python
latitude = []
longitude =[]

for i in addresses:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
restaurant['위도'] = latitude
restaurant['경도'] = longitude
restaurant.head(3)
```

```python
restaurant = restaurant[['소재지명', '도로명', '위도', '경도', '시설구분']]
restaurant.head(3)
```

```python
restaurant.to_csv('res/음식점.csv')
```

```python
hospital['소재지도로명주소'].str.split('(')[2]
```

```python
hospital['도로명'] = 0
for i in range(len(hospital)):
    hospital['도로명'][i] = hospital['소재지도로명주소'].str.split('(')[i][0].split(',')[0].rstrip()
hospital.head(3)
```

```python
hospital = hospital.rename(columns={'사업장명' : '소재지명'})
hospital.head(3)
```

```python
hospital_a = hospital[hospital['위도'].isnull()]
hospital_a
```

```python
hospital = hospital.dropna(axis=0)
hospital.head(3)
```

```python
addresses = []

for i in hospital_a.index:
  addresses.append(hospital_a['도로명'][i])

addresses
```

```python
latitude = []
longitude =[]

for i in addresses:
    latitude.append(geocoding(i)[0])
    longitude.append(geocoding(i)[1])
```

```python
hospital_a['위도'] = latitude
hospital_a['경도'] = longitude
hospital_a.head(3)
```

```python
hospital_all = pd.concat([hospital, hospital_a])
hospital_all
```
