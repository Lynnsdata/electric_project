```python
import pandas as pd
```

# 컬럼 rename

```python
# 관공서
office = pd.read_csv('./res/관공서현황.csv', index_col=0)
office.head()
```

```python
office.rename(columns={'소재지명':'place'}, inplace=True)
office.head()
```

```python
office_df = pd.read_csv('./res/관공서가중치BEST5.csv', index_col=0)
office_df.head()
```

```python
office_df.rename(columns={'office_name':'place'}, inplace=True)
office_df.head()
```

```python
office_df_merge = pd.merge(office, office_df, on='place', how='right')
office_df_merge.head()
```

```python
# 숙박업체
hotel = pd.read_csv('./res/숙박업소현황.csv', index_col=0)
hotel.head()
```

```python
hotel.rename(columns={'소재지명':'place'}, inplace=True)
hotel.head()
```

```python
hotel_df = pd.read_csv('./res/숙박가중치BEST5.csv', index_col=0)
hotel_df.head()
```

```python
hotel_df.rename(columns={'hotel_name':'place'}, inplace=True)
hotel_df.head()
```

```python
hotel_df_merge = pd.merge(hotel, hotel_df, on='place', how='right')
hotel_df_merge.head()
```

# 최종 concat

```python
df_score_all = pd.concat([office_df_merge, hotel_df_merge], ignore_index=True)
df_score_all
```

```python
df_score_all = df_score_all.drop_duplicates(['place'])
df_score_all
```

```python
df_score_all = df_score_all.sort_values(by='5_score' ,ascending=False)
df_score_all
```

```python
df_score_all.reset_index(drop=True, inplace=True)
df_score_all
```

```python
df_score_all.to_csv('./res/최종가중치_관공서_숙박업소.csv')
```
